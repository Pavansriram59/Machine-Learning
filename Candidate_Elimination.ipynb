```python
data=read_csv("C:\\Users\\exam2\\Downloads\\ENJOYSPORT.csv")
data
```

```python
#Implementing Candidate_Elimination Algo

att=len(data.columns)
specific_h=['@']*(att-1)
general_h=[["?" for i in range(len(specific_h))] for i in range(len(specific_h))]
print("Most specific hypothesis:\n",specific_h)
print("Most general hypothesis:\n",general_h)
print()

for i in range(len(data)):
    x=list(data.iloc[i,0:att])
    
    if(x[-1]==1):
        print("-------------------")
        print(f"Instance {i+1} is positive exapmle:")
        print(x[:att-1])
        print('-------------------')
        for j in range(len(x)-1):
            if specific_h[j]=='@' or specific_h[j]==x[j]:
                specific_h[j]=x[j]
            elif specific_h[j]!=x[j]:
                specific_h[j]='?'      
        print(f"\nAfter instance {i+1} the specific hypothesis is:\n",specific_h)
        print(f"\nAfter instance {i+1} the general hypothesis is:\n",general_h)
        print()
    else:
        print("-------------------")
        print(f"Instance {i+1} is negative exapmle:")
        print(x[:att-1])
        print('-------------------')
        for l in range(len(specific_h)):
            if x[l]!=specific_h[l]:
                general_h[l][l]=specific_h[l]
            else:
                general_h[l][l]='?'
        print(f"\nAfter instance {i+1} the specific hypothesis is:\n",specific_h)
        print(f"\nAfter instance {i+1} the general hypothesis is:\n",general_h)
        print()
```
